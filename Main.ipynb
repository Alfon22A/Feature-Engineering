## Imports

````python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)

from sklearn.impute import KNNImputer
````

````python
data = pd.read_csv("learningSet.csv")
data.head()
````

## Numerical NAs

````python
num = data.select_dtypes(np.number)
num_cols = pd.DataFrame(num.isna().sum()/len(num)).reset_index()
num_cols.columns = ["Cols", "NaNs"]
num_cols[num_cols["NaNs"] > 0]
````

## GEOCODE2

We consider the rows with " " as missing values and drop the rows with them.

````python
data["GEOCODE2"].value_counts(dropna = False)
````

````python
data["GEOCODE2"] = [x if x in ["A", "B", "C", "D"] else np.NaN for x in data["GEOCODE2"]]
````

````python
data["GEOCODE2"].value_counts(dropna = False)
````

````python
data = data[data["GEOCODE2"].notna()]
data["GEOCODE2"].value_counts(dropna = False)
````

## WEALTH1

There are way too many rows to drop them (almost 50% of the df) and this column is important, so we are going to use ```WEALTH2``` and ```INCOME``` assign it a value

````python
data["WEALTH1"].value_counts(dropna = False)
````

````python
data[["INCOME", "WEALTH1", "WEALTH2"]]
````

We drop those rows that had all 3 missing values.

````python
data = data.drop(index = data[data["INCOME"].isna() & data["WEALTH1"].isna() & data["WEALTH2"].isna()].index)
````

````python
data[["INCOME", "WEALTH1", "WEALTH2"]]
````

````python
data_wealth = data[["INCOME", "WEALTH1", "WEALTH2"]]
````

We assign ```INCOME``` to ```WEALTH2``` if it is missing and vicecersa.

````python
data_wealth[data_wealth["WEALTH2"].isna()]
````

````python
data_wealth["WEALTH2"][data_wealth["WEALTH2"].isna()] = data_wealth["INCOME"][data_wealth["WEALTH2"].isna()]
````

````python
data_wealth["WEALTH2"]
````

````python
data_wealth[data_wealth["INCOME"].isna()]
````

````python
data_wealth["INCOME"][data_wealth["INCOME"].isna()] = data_wealth["WEALTH2"][data_wealth["INCOME"].isna()]
````

````python
data_wealth["INCOME"]
````

And finally we assing the mean value as an int for those rows that have ```WEALTH1``` missing.

````python
data_wealth["WEALTH1"][data_wealth["WEALTH1"].isna()]
````

````python
data_wealth["WEALTH1"][data_wealth["WEALTH1"].isna()] = [int(np.mean([x, y])) for x, y in zip(data_wealth["INCOME"][data_wealth["WEALTH1"].isna()], data_wealth["WEALTH2"][data_wealth["WEALTH1"].isna()])]
````

````python
data_wealth = data_wealth.astype(int)
data_wealth
````

````python
data["WEALTH1"] = data_wealth["WEALTH1"]
````

````python
data = data.reset_index()
````

````python
data["WEALTH1"].isna().sum()
````

## ADI DMA MSA

As these 3 variables are location codes, we are going to use the KNN to guess the missing values (0)

````python
data_code = data[["ADI", "DMA", "MSA"]]
data_code.head()
````

````python
data_code.isna().sum()
````

````python
imputer_code = KNNImputer(missing_values = 0)
data_code_knn = imputer_code.fit_transform(data_code)
data_code_knn = pd.DataFrame(data_code, columns = ["ADI", "DMA", "MSA"])
print(data_code_knn["ADI"].value_counts())
print()
print(data_code_knn["DMA"].value_counts())
print()
print(data_code_knn["MSA"].value_counts())
````

````python
print(data["ADI"].value_counts())
print()
print(data["DMA"].value_counts())
print()
print(data["MSA"].value_counts())
````

````python
# As they are codes we make them integers
data_code_knn = data_code_knn.astype(int)
````

````python
print(data_code_knn["ADI"].value_counts())
print()
print(data_code_knn["DMA"].value_counts())
print()
print(data_code_knn["MSA"].value_counts())
````

````python
data[["ADI", "DMA", "MSA"]] = data_code_knn[["ADI", "DMA", "MSA"]]
````
